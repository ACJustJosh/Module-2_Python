```python
import csv
import pathlib

menu_csv_filepath = pathlib.Path("Resources/menu_data.csv")
sales_csv_filepath = pathlib.Path("Resources/sales_data.csv")

menu = []
with open(menu_csv_filepath) as menu_csv:
    reader = csv.reader(menu_csv)
    next(reader)  
    for row in reader:
        menu.append(row)

sales = []
with open(sales_csv_filepath) as sales_csv:
    reader = csv.reader(sales_csv)
    next(reader)  
    for row in reader:
        sales.append(row)

report = {}
```

```python
for row in sales:
    quantity = int(row[3])
    sales_item = row[4]

    if sales_item not in report:
        report[sales_item] = {
            "01-count": 0,
            "02-revenue": 0,
            "03-cogs": 0,
            "04-profit": 0,
        }

    for menu_row in menu:
        item = menu_row[0]
        price = float(menu_row[3])
        cost = float(menu_row[4])

        if sales_item == item:
            profit = price - cost
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += profit * quantity
        else:
            # Otherwise print error message
            print(f"{sales_item} does not equal {item}! NO MATCH!")
```

```python
output_file_path = pathlib.Path("Reports", "Report.txt")

with open(output_file_path, "w") as txt_file:
    for key, value in report.items():
        txt_file.write(f"{key} {value}\n")
```
